```python
import requests
import pprint
import json
import pandas as pd
url = "https://api.teleport.org/api/urban_areas/"

# Make a GET request to retrieve the data
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    data = response.json()
    # Process the data as needed
    
    city_names = [city['name'] for city in data['_links']['ua:item']]
    
    #pprint.pprint(data)
else:
    print("Request failed with status code:", response.status_code)
print(city_names)
```

```python
formatted_city_names=[]
for city_name in city_names:
    formatted_city_name = city_name.replace(" ", "-").replace(".","").lower()
    formatted_city_name = formatted_city_name.replace(",","")
    formatted_city_names.append(formatted_city_name)
formatted_city_names
print(len(formatted_city_names))
```

```python
city_scores = []

for city_name in city_names:
    formatted_city_name = city_name.replace(" ", "-").replace(".", "").lower()
    formatted_city_name = formatted_city_name.replace(",", "")
   
    url = f"https://api.teleport.org/api/urban_areas/slug:{formatted_city_name}/scores/"

    response = requests.get(url)

    if response.status_code == 200:
        city_data = response.json()
        
        category_scores = {'City': city_name}
        category_scores.update({category['name'] + " Score": category['score_out_of_10'] for category in city_data['categories']})
        category_scores['teleport_city_score'] = city_data['teleport_city_score']
        
        city_scores.append(category_scores)
    
    else:  
        if city_name == "Galway":
            url = f"https://api.teleport.org/api/urban_areas/slug:gaillimh/scores/"
            response = requests.get(url)
            
            if response.status_code == 200:
                city_data = response.json()
                
                category_scores = {'City': city_name}
                category_scores.update({category['name'] + " Score": category['score_out_of_10'] for category in city_data['categories']})
                category_scores['teleport_city_score'] = city_data['teleport_city_score']
        
                city_scores.append(category_scores)
        else:
            print(f"Request for {city_name}: {formatted_city_name} failed with status code:", response.status_code)

df = pd.DataFrame(city_scores)
df
```

```python
columns_to_keep = ['City', 'Cost of Living Score', 'Safety Score', 'teleport_city_score']


df = df[columns_to_keep]
df
```

```python
df.to_csv("city_scores.csv", index=False)
```

```python
from geopy.geocoders import Nominatim
name=[]
Lan=[]
Long=[]
for city_name in city_names:
    # Replace with the desired city name

    geolocator = Nominatim(user_agent="my-app")
    location = geolocator.geocode(city_name)

    if location is not None:

        latitude = location.latitude
        longitude = location.longitude
        name.append(city_name)
        Lan.append(latitude)
        Long.append(longitude)

    else:
        print("City {city_name} not found.")
City_location=pd.DataFrame({'City':name,'Latitude':Lan,'Longitude':Long})

display(City_location)
```
